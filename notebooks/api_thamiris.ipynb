```python
import os
from os.path import isfile, join
from os import listdir

pathh = ('../raw_data/SP') #caminho geral
files = [f for f in listdir(pathh) if isfile(join(pathh, f))] #lista de nomes de arquivos de dados

def get_data(n_cidade):#escolher um numero de 0 a 35
    #Loop para fazer lista com os dataframes, ignorando o cabeÃ§alho
    #Criando 4 novas features a partir de infos do cabeÃ§alho
    file = []
    cidade = cidades[n_cidade]
    for i in range(0, 63):#aqui eu seleciono somente os arquivos com barreto no nome
        if cidade in files[i]:
            file.append(files[i])

    df_list = [] #aqui eu crioo dataframe soh com a cidade selecionada
    for ii in range(0, len(file)):
        df = pd.read_csv(f'{pathh}/{file[ii]}', sep=';', skiprows=8, encoding="ISO-8859-1", decimal=',')
        lat_log_alt = pd.read_csv(f'{pathh}/{file[ii]}', sep=';', skiprows=4,
                        nrows=3, encoding="ISO-8859-1", decimal=',', names=['esta_codi_lat_lon_alt','valor'])
        df['EstaÃ§ao']=file[ii].split('_')[0]
       # df['Estacao']=lat_log_alt['valor'][0]
       # df['Codigo']=lat_log_alt['valor'][1]
        df['Latitude']=lat_log_alt['valor'][0]
        df['Longitude']=lat_log_alt['valor'][1]
        df['Altitude']=lat_log_alt['valor'][2]
        df_list.append(df)
    return df_list, lat_log_alt
```

```python
def get_lat_lon(n_files):
        codigoestacao, altitude, estacao, lat, lon = [], [], [], [], []
        for file in range(0,n_files):
            lat_lon_alt = pd.read_csv(f'{pathh}/{files[file]}', sep=';', skiprows=4,
                             nrows=3, encoding="ISO-8859-1", decimal=',', names=['lat_lon_alt','valor'])
            est=files[file].split('_')[4]
            latt=lat_lon_alt['valor'][0]
            lonn=lat_lon_alt['valor'][1]
            alti=lat_lon_alt['valor'][2]
            codigoest=files[file].split('_')[3]
            altitude.append(alti)
            estacao.append(est)
            lat.append(latt)
            lon.append(lonn)
            codigoestacao.append(codigoest)

        return lat, lon, estacao, altitude, codigoestacao
```

```python
alt_csv = get_lat_lon(63)
```

```python
df_alt = pd.DataFrame(alt_csv).T
```

```python
df_to_api = df_alt.drop_duplicates()
```
